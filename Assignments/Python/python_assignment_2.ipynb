```python
import pandas as pd
```

```python
employee_df = pd.read_csv('../files/employees.csv')
salaries_df = pd.read_csv('../files/salaries.csv')
department_df = pd.read_csv('../files/departments.csv')
projects_df = pd.read_csv('../files/projects.csv')
locations_df = pd.read_csv('../files/locations.csv')
```

1. Join employees.csv with salaries.csv on EmployeeID to get the salary details for each employee.

```python
# Implicit merge
merged_df = pd.merge(employee_df,salaries_df,how='inner')
merged_df
```

```python
# Explicit merge
merged_df = pd.merge(employee_df,salaries_df,on='EmployeeID',how='inner')
merged_df
```

Note:
| Situation                   | Columns Used for Join                          | Result        |
| --------------------------- | ---------------------------------------------- | ------------------------------------------------------------------------------------------------------ |
| **First Merge (Implicit)**  | All matching columns (`EmployeeID` & `Salary`) | Rows match exactly on both columns (`EmployeeID`, `Salary`)                                            |
| **Second Merge (Explicit)** | `EmployeeID` only                              | Rows match only on `EmployeeID`, thus differing salaries are shown separately (`Salary_x`, `Salary_y`) |


2. Left Join: Join employees.csv with departments.csv on Department to get the department manager for each employee.

```python
#employee row updated
#10,Susan,HRA,3000
leftjoin_df = pd.merge(employee_df,department_df,how='left',on='Department')
leftjoin_df
```

3. Right Join: Join employees.csv with projects.csv on EmployeeID to get the projects assigned to employees, including those with no project.

```python
# projects.csv --> 111,11,Project X
# employees.csv --> 12,Susan,HRA,3000
rightjoin_df = pd.merge(employee_df,projects_df,how='right',on='EmployeeID')
rightjoin_df
print(rightjoin_df)
```

4. Full Outer Join: Join salaries.csv with locations.csv on EmployeeID to get salary and location for all employees, even if some information is missing in one of the tables.

```python
fullouterjoin_df = pd.merge(salaries_df,locations_df,how='outer',on='EmployeeID')
fullouterjoin_df
```

5. Self Join: Join employees.csv with itself based on Department to get the list of employees in the same department as a specific employee.

```python
selfjoin_df = pd.merge(employee_df,employee_df,on='Department')
selfjoin_df
```
